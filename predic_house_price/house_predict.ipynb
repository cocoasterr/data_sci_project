```python
import pandas as pd

data = pd.read_csv('Bengaluru_House_Data.csv')

data = data.drop(['balcony', 'bath', 'society', 'availability'],axis='columns').dropna()

def convert_to_num (x):
    try:
        return float(x)
    except:
        try:
            x = x.split('-')
            return float(x[0]) / float(x[1])
        except:
            return None
data.total_sqft.unique()
data.total_sqft = data.total_sqft.apply(convert_to_num).dropna()
data.size = data['size'].apply(lambda x : str(x).split(' ')[0])
data.location = data.location.apply(lambda x : str(x).strip())
data.head()
```

```python
data.head()
location_group = data.groupby('location').location.agg('count').sort_values(ascending=False)
location_less_than_10 = location_group[location_group <= 10]
data.location = data.location.apply(lambda x : 'other' if x in location_less_than_10 else x)
len(data.location.unique())
data.head(15)
```

```python
from collections import Counter

def filter_unique(lst):
  return [item for item, count in Counter(lst).items() if count > 1]

filter_unique([1, 2, 2, 3, 4, 4, 5]) # [2, 4]
```
